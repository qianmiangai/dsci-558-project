```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
linkage_path = "../3_EntityLinking/artist_predictions.csv"
linkage_df = pd.read_csv(linkage_path, index_col=False)

artsy_path = "../2_Artsy/ArtsyResult/CleanedArtsyData.csv"
artsy_df = pd.read_csv(artsy_path)

ww_path = "../2_Widewalls/WidewallsData/CleanedWidewallsData.csv"
ww_df = pd.read_csv(ww_path)

artsy_name_path = "../2_Artsy/ArtsyResult/ArtsyArtist.csv"
artsy_name_df = pd.read_csv(artsy_name_path)
```

```python
linkage_list = []
for i in range(len(linkage_df)):
    linkage_list.append([linkage_df.iloc[i][0],linkage_df.iloc[i][1]])
artsy_list = [x[0] for x in linkage_list]
ww_list = [x[1] for x in linkage_list]
```

```python
header = list(artsy_df.columns)
header.append('biography')
print(header)
```

```python
artsy_name_df = artsy_name_df.loc[artsy_name_df['id'].isin(artsy_list)]
artist_list = list(artsy_name_df['artist_name'])
artsy_df = artsy_df.loc[artsy_df['artist_name'].isin(artist_list)]
ww_df = ww_df.loc[ww_df['id'].isin(ww_list)]
ww_df = ww_df.drop(['id'], axis = 1)
```

```python
artsy_df = artsy_df.merge(ww_df, on = "artist_name", how = 'left' )
artsy_df
```

```python
artsy_df.to_csv('FinalDataSet.csv', index=False)
```

```python
def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
```

```python
#shutil.copyfile(src, dst)
img_file_path = "../2_Artsy/ArtsyResult"
#print([x[1] for x in os.walk(img_file_path)])
img_folder_path = [x[2] for x in os.walk(img_file_path)]
img_path = []
for f in img_folder_path:
    img_path += f
final_img_path = []
for i in img_path:
    if i[:-4] in final_id:
        final_img_path.append(i)
print(len(final_img_path))
for  i in range(len(final_img_path)):
    final_img_path[i] = find(final_img_path[i],img_file_path)
```
