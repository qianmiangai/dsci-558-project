```python
import pandas as pd
import numpy as np
import os
import cv2
```

```python
img_folder = '../4_DataSetCreation/images'
img_height = 256
img_width = 256
input_shape = (img_height, img_width, 3) #3 for RGB/255
```

```python
def create_dataset_for_prediction_ID_Array(img_folder):
    c=0
    c2 = 0
    id_array=[]
    img_data_array=[]
    for file in os.listdir(img_folder):
        image_path= os.path.join(img_folder, file)
        image= cv2.imread( image_path, cv2.COLOR_BGR2RGB)
        try:
            image=cv2.resize(image, (img_height, img_width),interpolation = cv2.INTER_AREA)
        except:
            #print(file[:-4])
            c+=1
            continue
        image=np.array(image)
        image = image.astype('float32')
        image /= 255 
        if image.shape != input_shape:
            #print(file[:-4])
            c2+=1
            continue
        id_array.append(file[:-4])
        img_data_array.append(image)
    print(c,c2)
    return id_array, img_data_array
```

```python
id_list, raw_data = create_dataset_for_prediction_ID_Array(img_folder)
```

```python
len(id_list), len(raw_data), len(raw_data[0][0][0]),raw_data[0].shape
```

```python
data_tuples = list(zip(id_list,raw_data))
df = pd.DataFrame(data_tuples, columns=['id', 'data_array'])
df.head()
```

```python
df.to_csv('RGB_array.csv')
```
