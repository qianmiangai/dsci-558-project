```python
artist_styles = df.groupby('artist_name')['class'].apply(list)
```

```python
artist_styles.head()
```

```python
type(artist_styles['Julio Le Parc'][0])
```

```python
artists_most_common_style = []
for artist, styles in zip(artist_styles.index, artist_styles):
    c = Counter(styles)
    c = Counter({k:v for k, v in c.items() if isinstance(k, str)})
    try:
        most_common_style = list(c.most_common())[0][0]
    except:
        most_common_style = "no distinct style"
#     print(c)
#     break
    artists_most_common_style.append([artist, most_common_style])
```

```python
artist_style_df = pd.DataFrame(artists_most_common_style, columns=['artist_name', 'artist_style'])
```

```python
len(artist_style_df)
```

```python
artist_style_df.head()
```

```python
df = df.merge(artist_style_df, on='artist_name')
```
